```python
# For running Spark
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName("Spark").getOrCreate()# For running Pandas on top of Spark
import pyspark.pandas as ps
```

```python
# SPARK
sdf = spark.read.options(inferSchema='True',
              header='True').csv('iris.csv')
sdf.show(5)
```

```python
# PANDAS-ON-SPARK
pdf = ps.read_csv('iris.csv')
```

```python
pdf.head()
```

```python
#SELECT

# SPARK
sdf.select("SepallengthCm","SepalWidthCm").show()

# PANDAS-ON-SPARK
pdf[["SepalLengthCm","SepalWidthCm"]].head()
```

```python
#DROP COLUMNS

# SPARK
sdf.drop('SepalLengthCm').show()

# PANDAS-ON-SPARK
pdf.drop('SepalLengthCm').head()
```

```python
#DROP DUPLICATES

# SPARK
sdf.dropDuplicates(["SepalLengthCm","SepalWidthCm"]).show()

# PANDAS-ON-SPARK
pdf[["SepalLengthCm", "SepalWidthCm"]].drop_duplicates()
```

```python
#FILTER

# SPARK
sdf.filter( (sdf.Species == "Iris-setosa") & (sdf.PetalLengthCm > 1.5) ).show()

# PANDAS-ON-SPARK
pdf.loc[ (pdf.Species == "Iris-setosa") & (pdf.PetalLengthCm > 1.5) ].head()
```

```python
#COUNT

# SPARK
sdf.filter(sdf.Species == "Iris-virginica").count()


```

```python
# PANDAS-ON-SPARK
pdf.loc[pdf.Species == "Iris-virginica"].count()
```

```python
#DISTINCT

# SPARK
sdf.select("Species").distinct().show()

# PANDAS-ON-SPARK
pdf["Species"].unique()
```

```python
#SORT

# SPARK
sdf.sort("SepalLengthCm", "SepalWidthCm").show()

# PANDAS-ON-SPARK
pdf.sort_values(["SepalLengthCm", "SepalWidthCm"]).head()
```

```python
#GROUP BY

# SPARK
sdf.groupBy("Species").count().show()

# PANDAS-ON-SPARK
pdf.groupby("Species").count()
```

```python
#REPLACE

# SPARK
sdf.replace("Iris-setosa", "setosa").show()

# PANDAS-ON-SPARK
pdf.replace("Iris-setosa", "setosa").head()
```

```python
#JOIN

# SPARK
sdf.union(sdf).show()

# PANDAS-ON-SPARK
pdf.append(pdf)
```
